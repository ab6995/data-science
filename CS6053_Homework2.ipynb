2\. Write a Python function called `getDfSummary()` that does the following:
- Takes as input a data frame
- For each variable in the data frame calculates the following features:
  - `number_nan` to count the number of missing not-a-number values
  - Ignoring missing, NA, and Null values:
    - `number_distinct` to count the number of distinct values a variable can take on
    - `mean`, `max`, `min`, `std` (standard deviation), and `25%`, `50%`, `75%` to correspond to the appropriate percentiles
- All of these new features should be loaded in a new data frame. Each row of the data frame should be a variable from the input data frame, and the columns should be the new summary features.
- Returns this new data frame containing all of the summary information

Hint: The pandas `describe()` [(manual page)](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.describe.html) method returns a useful series of values that can be used here.

```python
def getDfSummary(input_data):
    #Initialize the lists to store the computed data
    ads = input_data
    attribute_name = [] #For storing all names 
    number_nan = [] #number of missing not-a-number values
    number_distinct = [] #number of distinct values
    mean = []  #mean
    max_num = [] #max
    min_num = [] #min
    std_dev = [] #std
    quantile25 = []#25% percentile
    quantile50 = []#50% percentile
    quantile75 = []#75% percentile
    #just filling out NAN with '0' for calculating percentile part as it was showing me runtime warning 
    ads1 = input_data.fillna(0)
    
    #For looping through the columns of data frame 
    for column in input_data.columns:
        attribute_name.append(column)
        number_nan.append(input_data[column].isnull().sum())
        number_distinct.append(input_data[column].nunique())
        mean.append(input_data[column].mean())
        max_num.append(input_data[column].max())
        min_num.append(input_data[column].min())
        std_dev.append(input_data[column].std())
        quantile25.append(input_data[column].quantile(0.25))#quantile25
        quantile50.append(input_data[column].quantile(0.50))#quantile50
        quantile75.append(input_data[column].quantile(0.75))#quantile75
    
    #Create output data frame and reaaranging the data according to answer
    output_data = pd.DataFrame({"variable":attribute_name,"number_nan":number_nan, "number_distinct":number_distinct, "mean":mean, "max":max_num, "min":min_num, "std":std_dev, "25%":quantile25, "50%":quantile50, "75%":quantile75})
    output_data = output_data[['variable', 'number_nan', 'number_distinct', 'mean', 'max', 'min', 'std', '25%', '50%', '75%']]
        
    return output_data
#Calling
getDfSummary(ads) 
```
